### Building the master dataframe

```python
import pandas as pd


all_df = pd.read_csv('../IAQF_DATA.csv')

rates_df = pd.DataFrame()
rates_df['Dates'] = all_df['Dates_rfrate']
rates_df['rf_rate'] = all_df['rf_rate']

del all_df['Dates_rfrate']
del all_df['rf_rate']

master = pd.merge(all_df, rates_df, how='inner', on=['Dates'])
```

```python
from scipy.stats import norm
from math import log, sqrt, exp

def d_plus( x, K, tau, sigma, rfr ):
    
    return ( (log(x/K) + ((rfr + 0.5*(sigma**2))*tau))/(sigma*sqrt(tau)) )

def d_minus( x, K, tau, sigma, rfr ):
    
    return ( d_plus(x, K, tau, sigma, rfr) - sigma*sqrt(tau) )

def bs_call_price( x, K, tau, sigma, rfr ):
    
    tau /= 360
    rfr /= 100
    sigma /= 100

    d_1 = d_plus(x, K, tau, sigma, rfr)
    d_2 = d_minus(x, K, tau, sigma, rfr)
    
    return ( x*norm.cdf(d_1) - K*exp(-1.0*rfr*tau)*norm.cdf(d_2) )

def bs_put_price( x, K, tau, sigma, rfr ):
    
    tau /= 360
    rfr /= 100
    sigma /= 100

    d_1 = d_plus(x, K, tau, sigma, rfr)
    d_2 = d_minus(x, K, tau, sigma, rfr)
    
    return ( K*exp(-1.0*rfr*tau)*norm.cdf(-1.0*d_2) - x*norm.cdf(-1.0*d_1) )
```

Add in Black-Scholes call and put prices, along with the 60, 120 day moving averages:

```python
import numpy as np

master.index = pd.RangeIndex(start=1, stop=3249)

master['bs_call'] = np.vectorize(bs_call_price)(master.SP500, master.SP500, 90, master.vol, master.rf_rate)
master['bs_put'] = np.vectorize(bs_put_price)(master.SP500, master.SP500, 90, master.vol, master.rf_rate)
master['MA_60']  = master.SP500.rolling(window=60, center=False).mean()
master['MA_120'] = master.SP500.rolling(window=120, center=False).mean()
```

```python
"""
The daily rebalance include using the delta
and gamma of the portfolio. Hence, we need functions to include
these metrics in our master dataframe.

This will all be moved up to the beginning so we can slice-n-dice
it for each portoflio. 
"""

# BS Call Delta
def call_delta( x, K, tau, sigma, rfr ):
    
    tau /= 360
    rfr /= 100
    sigma /= 100
    
    return ( norm.cdf(d_plus(x, K, tau, sigma, rfr)) )

# BS Put Delta
def put_delta( call_delta ):
    
    return ( call_delta - 1.0 )

# BS Call/Put Gamma (Equal)
def bs_gamma( x, K, tau, sigma, rfr ):
    
    tau /= 360
    rfr /= 100
    sigma /= 100
    
    d_1    = d_plus(x, K, tau, sigma, rfr)
    d_1_sq = d_1**2
    
    return ( (1.0/(x*sigma*sqrt(tau)))*(1/sqrt(2*np.pi))*np.exp(-0.5*d_1_sq) )

#Add the call and put deltas for each day
master['call_delta'] = np.vectorize(call_delta)(master.SP500, master.SP500, 90, master.vol, master.rf_rate)
master['put_delta']  = np.vectorize(put_delta)(master.call_delta)
master['bs_gamma'] = np.vectorize(bs_gamma)(master.SP500, master.SP500, 90, master.vol, master.rf_rate)

master.head()
```

```python
p4 = master.drop(['MA_60','MA_120','cross_over'], axis=1)

p4['bs_strad']     = p4.bs_call + p4.bs_put         # Define the straddle
p4['port_noCP'] = 0.0
p4['port_noAsset']  = 0.0
p4['port_delta']   = p4.call_delta + p4.put_delta   # Define portfolio_delta
p4['straddle_gamma']   = 2.0*p4.bs_gamma                # Define portfolio_gamma
                        
p4.head()
```
